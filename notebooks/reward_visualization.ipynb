```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load Simulated Dataset
df = pd.read_csv("../data/simulated_logs.csv")
df.head()
```

```python
# Step 3: Basic Info & Class Distribution
print("\nClass Distribution:")
print(df['label'].value_counts())
```

```python
plt.figure(figsize=(6, 4))
df['label'].value_counts().plot(kind='pie', labels=['Normal (0)', 'Attack (1)'], autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff6666'])
plt.title("Normal vs Attack Samples")
plt.ylabel("")
plt.show()
```

```python
# Step 4: Feature Distributions
features = ['login_attempts', 'cpu_usage', 'port_scans', 'file_changes']
for feature in features:
    plt.figure(figsize=(7, 4))
    sns.histplot(data=df, x=feature, hue='label', kde=True, palette='Set1', bins=30)
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.legend(title='Label', labels=['Normal', 'Attack'])
    plt.tight_layout()
    plt.show()
```
